O arquivo avocado.csv consiste em um dataset que contém dados sobre vendas de abacates (avocado, em inglês) em diversas regiões dos Estados Unidos. Essa dataset contém as seguintes colunas:

O arquivo googleplaystore.csv é um dataset que contém diversas informações sobre aplicativos disponíveis para download na Google Play Store. Logo após, verifique se existem dados duplicados e/ou ausentes nesse dataset.

#### Caso existam dados duplicados, remova-os. Caso existam dados ausentes, você pode removê-los ou pensar em alguma maneira de preenchê-los.


```python
df_g = pd.read_csv('aplgoogle.csv')
df_g
```

```python
df_g.info()
```

```python
# contando o máximo de nulos no dataframe.
max(df_g.isna().sum())

```

```python
# Transformando todos os Rating null em 0.0
df_g['Rating'].fillna(0, inplace=True)
df_g['Rating'].isna().sum()
```

```python
df_g['Type'].isna().sum()
```

```python
df_g['Type'].fillna('Free', inplace=True)
df_g['Type'].isna().sum()
```

```python
df_g['Content Rating'].isna().sum()
```

```python
df_g['Content Rating'].fillna('Mature 17+', inplace=True)
df_g['Content Rating'].isna().sum()
```

```python
df_g['Current Ver'].fillna('Varies with device', inplace = True)
df_g['Current Ver'].isna().sum()
```

```python
max(df_g.isna().sum())
```

```python
df_g['Android Ver'].isna().sum()
```

```python
df_g['Android Ver'].fillna('Varies with device', inplace=True)
df_g['Android Ver'].isna().sum()
```

```python
df_g.info()
```

```python
df_g['App'].duplicated().sum()
```

```python
df_g.drop_duplicates(keep='first', inplace=False)
```
